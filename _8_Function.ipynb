```python
#Function (Lambda)
# x -> x^2
square_formula = lambda x: x **2
print (square_formula(10))

#Python 
ages = [7, 60, 42, 24, 67, 80]
elderly_ages = list(filter(lambda x: x > 65, ages))
print(elderly_ages)

names = ["John" , "Jennie", "Tommmy", "Lucas", "Leo"]
#name starts J and length >5 -->list
new_names2 = list(filter(lambda name: name.startswith('J') and len(name)>5 , names))
print(new_names2)
```
