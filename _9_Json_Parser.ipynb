```python
# Serialization / Deserialization

import json

#python:JSON String -> Dictionary
json_str ='{"name": "Leo", "age":17}'
dict1 = json.loads(json_str)
print(type(dict1))
print(dict1)

print(dict1['name'])
```

```python
#read person.json
# find country Frnace and ages >25 ->list
with open("person.json") as file:
  person_list = json.load(file)
  new_person_list = list(filter(lambda p: p['age']>25 and p['address']['Country'] == 'France', person_list))
  str_list = json.dumps(new_person_list, indent=2)
  print(str_list)
```
