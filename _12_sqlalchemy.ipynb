```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import Table, Column, MetaData, Integer, String, Float, Date, BigInteger

# Part 1: create table

db_engine = create_engine("postgresql+psycopg2://postgres:admin1234@localhost:5432/bootcamp")

metadata = MetaData()

person_schema = Table(
  "persons",
  metadata,
  Column("id", BigInteger, primary_key=True),
  Column("name", String(50), nullable=False),
  Column("dob", Date, nullable=False),
  Column("salary", Float, nullable=False)
)

metadata.create_all(db_engine)
```

```python
# Part 2
df_person = pd.read_csv("persons.csv")
print(df_person.head())
print(df_person.info())

# ETL (Extract Transform Load)
df_person['dob'] = pd.to_datetime(df_person['dob'].astype(str), format="%Y-%m-%d")

# insert into db
df_person.to_sql("persons", db_engine, index=False, if_exists="replace")
```

```python
#DB select

db_sql_result = pd.read_sql("select * from persons where salary > 20000", con=db_engine)
print(db_sql_result)
```
