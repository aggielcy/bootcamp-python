```python
# Data Science
import pandas as pd

# Series (column)
ages = pd.Series([2, 8, 9, 40])
print(type(ages))

staff = {
  "name" : "Leo",
  "age" : 17
}

# DataFrame (table)
df = pd.DataFrame(staff, index=[0])
print(df)

# Dictionary -> { : , : }
# Set -> { , }
team = {
  "names" : ["Leo", "Lucas", "Vincent"],
  "ages" : [17, 14, 19]
}

print(type(team))
df_team = pd.DataFrame(team)
print(df_team)
```

```python
print(type(df_team["ages"]))

# loc -> series true/false -> a column of true/ false

df_new_team = df_team.loc[df_team["ages"] > 15]
print(df_new_team)

# name startswith L
df_new_team = df_team.loc[df_team["names"].str.startswith("L")]
print(df_new_team)

# and or
# age > 18 and name contain e
df_new_team = df_team.loc[(df_team["names"].str.contains("e")) & (df_team["ages"] > 18)]
print(df_new_team)

# name len > 5
df_new_team = df_team.loc[df_team["names"].str.len() > 5]
print(df_new_team)
```

```python
print (df_team.info())

print (df_team.describe())

print (df_team.head())
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

# Pandas -> read csv format to df
df = pd.read_csv(url)
print(type(df))

print(df.head())
print(df.describe())
```

```python
# Sex -> Female and age > 40
df_new = df.loc[(df['Sex'] == 'female') & (df['Age'] > 40)]
print(df_new.head())
print(df_new.describe())
```

```python
print(df['Age'].isnull().any())

print(df['Age'].isnull()) # return series

print(df.loc[df['Age'].isnull()])
```

```python
# Drop Data with null age
df_age_drop_na = df.dropna(subset=['Age'])
print(df_age_drop_na)
print(len(df_age_drop_na)) # 714

print(df_age_drop_na['Age'].isnull().any()) # False
```

```python
# Refill
df_for_refill = df.copy()
df_for_refill['Age'] = df_for_refill['Age'].fillna('N/A') # Overwrite

print(df_for_refill['Age'].isnull().any()) # False
print(len(df_for_refill))
```

```python
df_result = df.groupby('Sex')['Age'].mean()
print(df_result)

df_result = df.groupby(['Survived', 'Sex'])['Age'].max()
print(df_result)
```
