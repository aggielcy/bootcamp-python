```python
#Data Science
import pandas as pd

#Series(column)
ages = pd.Series([2,8,9,40])
print(type(ages))

staff ={
  "name" :"Leo",
  "age": 17
}

#DataFrame (Table)
df = pd.DataFrame(staff, index=[0])
print (df)

#Dictionary -> { : , : }
#set { , }
team = {
  "name" : ["Leo", "Lucas","Vincent"],
  "ages" : [17,14,19]
}

print (type (team))
df_team = pd.DataFrame(team)
print(df_team)

```

```python
print(type(df_team["ages"]))
#loc -> series true/false -> a column of true/ false
# it is like a SQL 

df_new_team = df_team.loc[df_team["ages"] >15]
print(df_new_team)

# name startsWith
df_new_team = df_team.loc[df_team["names"].str.startswith("L")]
print [df_new_team]

#and or
#age >15 and name startswith L
df_new_team = df_team.loc[(df_team["names"].str.contains("E")) & (df_team["ages"]>15)]
print [df_new_team]

# name len >15
df_new_team = df_team.loc[df_team["names"].str.len() >5]
print [df_new_team]

```

```python
print (df_team.info())

print (df_team.describe())

print (df_team.head())
```
