```python
# list
language = ['Java' , 'Python', 'C++']

language.append('C#')
print (language)


# remove, insert, sort, pop, clear, len
# list 1 + list 2
language.remove('Java')
print (language) #['Python', 'C++', 'C#']

language.insert(0,'apple')
print (language) #['apple', 'Python', 'C++', 'C#']

language.sort()
print (language) # ['C#', 'C++', 'Python', 'apple']

pop_result = language.pop(2) #remove as well
print (pop_result) # Python

print(len(language)) #3

human_language = ['English','Chinese']

languages = language +human_language
print(languages)
```

```python
# dictionary (Java: map)

person = {
  "name":"John" ,
  "age": 17
}

print (type(person))
print(person['name']) #similar to Java key to get value
print(person['age'])


trash = {
  17: "steven",
  0.99 : "Trump",
  True : False
}
print (trash[0.99])  #Trump


print(trash.keys()) #dict_keys([17, 0.99, True])
print(trash.values()) #dict_values(['steven', 'Trump', False])

for k,v in trash.items():
  print(f"key={k}, value={v}")

person ['age'] = 100
print(person)

person ['age'] = [100,200,300]  #<--demonstrate crazy explose possibility in python
print(person)
```
