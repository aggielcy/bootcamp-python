```python
import numpy as np

arr = np.array([20, 2, 10, 9, 2])
print (type(arr))

print(arr +1)
print(arr *2)

print(np.sum(arr))
print(np.mean(arr))
print(np.median(arr))
print(np.std(arr))
print(np.var(arr))
print(np.min(arr))
print(np.max(arr))

list1 = [20, 2, 10, 9]
print(list1*2)

print(np.sort(arr))

print(np.unique(arr))
```
